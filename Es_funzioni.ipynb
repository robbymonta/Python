```python
def posneg(x):
  if x <0:
    return "il  num è neg"
  elif x == 0:
    return "il num è zero, cambiare numero"
  else:
    return "Il num è pos"
x = float(input("Inserisci un numero: "))
print(posneg(x))
```

```python
def pd(x):
  if x %2 == 0:
    return "il numero è pari"
  else: 
    return "il numero è dispari"

x = int(input("Inserisci un numero: "))

print(pd(x))
```

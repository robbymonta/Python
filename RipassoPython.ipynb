```python
#Es6
#Ogni anno in banca sui conti deposito si accumula un interesse rispetto al saldo iniziale, l’interesse accumulato si somma al saldo precedente e concorrerà al calcolo dell’interesse sull’anno successivo.
#Dato un saldo iniziale, la percentuale di interesse e una soglia, verificare  dopo quanti anni si raggiunge o supera la soglia.
inte=float(input("Inserisci un tasso di interesse: "))
si=float(input("Inserisci un saldo iniziale: "))
sf=float(input("Inserisci un saldo finale: "))
def cont(si,int,sf):
  c=0
  ns=si+(si*inte/100)
  while ns <sf:
    ns=si+(si*inte/100)
    si=ns
    c+=1
  return (c)
print("la soglia di",sf,"viene raggiunta in",cont(si,inte,sf),"anni")
```

```python
#dati due numeri naturali (diversi da zero), trovare il loro minimo comune multiplo
def mcm(x,y):
  if x> y:
    greater =x
  else:
    greater =y
  while(True):
    if((greater % x==0) and (greater % y==0)):
      mcm =greater
      break
    greater +=1
  return mcm

num1 =float(input("Inserisci un numero: "))
num2 =float(input("Inserisci un numero: "))
print("Il minimo comune multiplo è: ",mcm(num1,num2))
```

```python
#massimo comune divisore
def mcd(a,b):
  while b:
    a,b=b,a%b
  return a 
num1 =float(input("Inserisci un numero: "))
num2 =float(input("Inserisci un numero: "))
print("Il massimo comune divisore è: ",mcd(num1,num2))
```

```python
#numero primo
numero=float(input("Inserire un numero: "))
if numero<=1:
  print("Devi inserire un numero maggiore di 1")
else:
  div,count=2,0
  while div<=numero/2 and count==0:
    if numero%div==0:
      count+=1
    div+=1
  if count==0:
    print("Numero primo!")
  else:
    print("Il numero non è primo!")
```

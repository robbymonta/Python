###MATPLOT

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(-5,5,100)
y= 2*x+1
plt.figure(figsize=(15,10))
plt.plot(x,y,"-r", label ="y=2x+1")
plt.title("Graph of y= 2x+1")
plt.xlabel("x",color ="#1c2843")
plt.ylabel("y",color ="#1c2843")
plt.legend(loc="upper left")
plt.grid()
plt.show()
```

```python
x = np.linspace(-5,5,100)
#the function, 
y=x**3
fig=plt.figure(figsize=(10,10))
ax=fig.add_subplot(1,1,1)
ax.spines["left"].set_position("center")
ax.spines["bottom"].set_position("center")
ax.spines["right"].set_position("center")
ax.spines["top"].set_position("center")
ax.xaxis.set_ticks_position("bottom")
ax.yaxis.set_ticks_position("left")
plt.plot(x,y, "g")
plt.show
```

```python
# 100 linearly spaced numbers
x = np.linspace(-2*np.pi,2*np.pi,100)
# the functions, which are y = ? and z = ? here
y = np.sin(x)
z = np.cos(x)
# setting the axes at the centre
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
# plot the functions
plt.plot(x,y, 'c', label='y=sin(x)')
plt.plot(x,z, 'm', label='y=cos(x)')
plt.legend(loc='upper left')
# show the plot
```

```python
# 100 linearly spaced numbers
x = np.linspace(-2,2,100)
# the function, which is y = e^x here
y = np.exp(x)
# setting the axes at the centre
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
# plot the function
plt.plot(x,y, 'y', label='y=e^x')
plt.legend(loc='upper left')
# show the plot
plt.show()
```
